```python
%%R

library(ggplot2)
library(dplyr)

# Load and filter the dataset
df_long <- read.csv("maritime_summary_long_yearly.csv") %>% 
    filter(year <= 2024)

# Create a highlight column for Panama Canal
df_long <- df_long %>%
    mutate(line_size = ifelse(portname == "Panama Canal", 2, 0.7),  # Extra bold for Panama
           line_color = ifelse(portname == "Panama Canal", "Panama Canal", portname))  # Keep original names

# Define colors: Panama = red, Bab el-Mandeb = manually assigned new color
other_ports <- setdiff(unique(df_long$portname), c("Panama Canal", "Bab el-Mandeb Strait"))  # Exclude these two from auto colors
other_colors <- rainbow(length(other_ports))  # Assign unique colors
custom_colors <- c("Panama Canal" = "red", 
                   "Bab el-Mandeb Strait" = "darkorange",  # Assign a distinct non-red color
                   setNames(other_colors, other_ports))  # Assign all other ports unique colors

# Plot with Panama Canal in red and Bab el-Mandeb Strait corrected
ggplot(df_long, aes(x = year, y = traffic, group = portname, color = line_color)) +
    geom_line(aes(size = line_size), alpha = 0.9) +  # Thicker Panama line
    geom_point(alpha = 0.9) +  # Points for all ports
    scale_size_identity() +  # Use the predefined size mapping
    scale_color_manual(values = custom_colors, 
                       breaks = c("Panama Canal", "Bab el-Mandeb Strait", other_ports)) +  
    labs(title = "Maritime Traffic Trends (2019-2024)",
         x = "Year", y = "Traffic Volume", color = "Port Name") +  # Proper legend label
    theme_minimal()
```

```python
%%R

library(ggplot2)
library(dplyr)

# Load and filter the dataset
df_long <- read.csv("maritime_summary_long_yearly.csv") %>% 
    filter(year <= 2024)

# Create a highlight column for Panama Canal
df_long <- df_long %>%
    mutate(line_size = ifelse(portname == "Panama Canal", 2, 0.7),  # Extra bold for Panama
           line_color = ifelse(portname == "Panama Canal", "Panama Canal", portname))  # Keep original names

# Define custom color palette WITHOUT red tones
safe_colors <- c(
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", 
    "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"
)  # Excludes bright reds
other_ports <- setdiff(unique(df_long$portname), "Panama Canal")  # List of non-Panama ports
custom_colors <- c("Panama Canal" = "red", setNames(safe_colors[1:length(other_ports)], other_ports))

# Plot with Panama Canal in red and all other ports in non-red colors
ggplot(df_long, aes(x = year, y = traffic, group = portname, color = line_color)) +
    geom_line(aes(size = line_size), alpha = 0.9) +  # Thicker Panama line
    geom_point(alpha = 0.9) +  # Points for all ports
    scale_size_identity() +  # Use the predefined size mapping
    scale_color_manual(values = custom_colors, 
                       breaks = c("Panama Canal", other_ports)) +  
    labs(title = "Maritime Traffic Trends (2019-2024)",
         x = "Year", y = "Traffic Volume", color = "Port Name") +  # Proper legend label
    theme_minimal()
```

```python
%%R

library(ggplot2)
library(dplyr)

# Load and filter the dataset
df_long <- read.csv("maritime_summary_long_yearly.csv") %>% 
    filter(year <= 2024)

# Create a highlight column for Panama Canal
df_long <- df_long %>%
    mutate(line_size = ifelse(portname == "Panama Canal", 2, 0.7),  # Extra bold for Panama
           line_color = ifelse(portname == "Panama Canal", "Panama Canal", portname))  # Keep original names

# Generate distinct colors using a high-contrast color palette
library(RColorBrewer)
color_palette <- brewer.pal(n = min(9, length(unique(df_long$portname))), name = "Set1")  # High contrast
other_ports <- setdiff(unique(df_long$portname), "Panama Canal")  # List of non-Panama ports
custom_colors <- c("Panama Canal" = "red", setNames(rep(color_palette, length.out = length(other_ports)), other_ports))

# Plot with Panama Canal in red and all other ports in distinct colors
ggplot(df_long, aes(x = year, y = traffic, group = portname, color = line_color)) +
    geom_line(aes(size = line_size), alpha = 0.9) +  # Thicker Panama line
    geom_point(alpha = 0.9) +  # Points for all ports
    scale_size_identity() +  # Use the predefined size mapping
    scale_color_manual(values = custom_colors, 
                       breaks = c("Panama Canal", other_ports)) +  
    labs(title = "Maritime Traffic Global Maritime Hotspots(2019-2024)",
         x = "Year", y = "Traffic Volume (Amount of ships passing the Chokepoint)", color = "The World’s 24 Chokepoints") +  # Proper legend label
    theme_minimal()
```
